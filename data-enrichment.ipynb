```python
import pandas as pd
from matplotlib import pyplot as plt 
```

```python
asteroids_df = pd.read_pickle("data/lvl1/asteroids_merged.pkl")
```

```python
# The Bus Class used in the SMASS data has too many classes for us to classify with machine learning
# We can map the various classes to the four classes in main group using this dictionary 
bus_to_main_dict = {
                    'A': 'Other',
                    'B': 'C',
                    'C': 'C',
                    'Cb': 'C',
                    'Cg': 'C',
                    'Cgh': 'C',
                    'Ch': 'C',
                    'D': 'Other',
                    'K': 'Other',
                    'L': 'Other',
                    'Ld': 'Other',
                    'O': 'Other',
                    'R': 'Other',
                    'S': 'S',
                    'Sa': 'S',
                    'Sk': 'S',
                    'Sl': 'S',
                    'Sq': 'S',
                    'Sr': 'S',
                    'T': 'Other',
                    'V': 'Other',
                    'X': 'X',
                    'Xc': 'X',
                    'Xe': 'X',
                    'Xk': 'X'
                   }
```

```python
# We now map the bus class to the main class and can drop some unnecessary columns to obtain our final dataframe
asteroids_df.loc[:, "Main_Group"] = asteroids_df["Bus_Class"].apply(lambda x: bus_to_main_dict.get(x, "None"))
asteroids_df.drop(columns=["DesNr", "FilePath"], inplace=True)
# print(asteroids_df)
```

```python
plt.plot(asteroids_df.iloc[0]["SpectrumDF"]["Wavelength_in_microm"], asteroids_df.iloc[0]["SpectrumDF"]["Reflectance_norm550nm"])
```

```python
# Let's save the final dataframe in a lvl2 data folder
asteroids_df.to_pickle("data/lvl2/asteroids.pkl")
```
